Is it possible to use Rust in jupyter notebook as well as python? Of course it is possible.  
The benefits are, for example, the ability to run code in sections to see data, and the ease   
of inserting markdown notes, which is very comfortable when learning code,   
and most importantly, the ability to integrate it into VSCode.  

This blog will cover how to install Jupyter Notebook on windows system and how to connect Rust kernel in Jupyter Notebook.

### **Step 1: Install Jupyter Notebook**
- If you already have Jupyter Notebook installed, you can skip this step

1. First we need to install Python, because Jupyter Notebook is based on Python.
   - If you already have Python installed, you can skip this step
   - Install Python[Official Download](https://www.python.org/downloads/)

2. Install Jupyter Notebook via pip
   - `pip3 install jupyter`

### **Step 2: Install the Rust kernel**
1. Assuming you already have Rust installed locally, if not, you can refer to[Official Download](https://www.rust-lang.org/tools/install)
2. Enter the following command in bash
   1. `cargo install evcxr_jupyter`
   2. `evcxr_jupyter --install`
3. You will also need to install the source code for the Rust standard library. If you are already using rust-analyzer,   
you probably already have it installed. To install it using rustup, run`rustup component add rust-src`

### **Step 3: Install the Vscode Extensions**
1. **Jupyter**   
   
2. **Python**  
  

### **Step 4: Create a blank Jupyter Notebook in Vscode**
1. First restart your VSCode
2. Press the shortcut key in VSCode `Ctrl+Shift+P`enter `create New Jupyter Notebook`.   
   
3. At this point we will get a blank .ipynb file, click `Select Kernel` in the upper right corner and select `Jupyter Kernel` -> `Rust`  
      

      

      
   
4. Congratulations! At this point we're ready to write Rust code in Jupyter Notebook!    
     

### **Usage Tips**
1. In Jupyter Notebook there is no need to write the function fn main(), we just need to write the code directly in the code block.  
   
2. The kernel evcxr_jupyter does not support reading input directly from stdin, so we need to use `evcxr_input` to read input 
       
     
3. Importing external libraries can be a bit tricky, but you can do it like this  
     
4. You can just take a screenshot with another software and then `Ctrl+V` paste it directly into MarkDown.   
     
   
5. You can drag and drop files directly from the Explorer on the left side of Vscode into MarkDown, hold down the `Shift` key and let go to quickly create a clickable link label. if you drag and drop into the code block you can quickly get the file path.
6. Click on "..." in the bar at the top Select "Export" to quickly export to html or pdf files for easy sharing. In fact, this blog was exported using this function.   
      
      
7. For more tips and tricks, please refer to the [official documentation](https://github.com/evcxr/evcxr/blob/main/COMMON.md)
