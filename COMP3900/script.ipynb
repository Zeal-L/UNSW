The backend of our project is using Flask framework. Development is done in Python language.

I have divided the whole backend into three parts which are:
- models: used to store the database model
- services: used to store the business logic
- routes: used to store all the routes

- models:
    - The database we use is PostgreSQL, which can support a lot of features, such as lists as fields, json as fields, and even custom data types.

    - This diagram shows the ER diagram of our database. You can see from the colors that our database has four main parts.
    - These tables are related by foreign keys, for example the relationship between the `vouchers` table and the `voucherTemplate` table is through the `template_id` foreign key. This way, vouchers only need to store information about whether they are used or not and who owns them, and the specific information about the coupon is stored in the voucherTemplate.
    - This means that vouchers of the same type can point to the same template, which reduces database redundancy significantly.

    - By the way, user passwords are not stored in clear text, but encrypted.

    - And then the actual model code can take the table dishes as an example
    - You can see that each of our tables is a class that inherits from the `db.Model` class, and then it's just a matter of writing the table's fields inside the class. The library sqlalchemy will help us to serialize and deserialize the objects automatically.
    - We can also create some static functions to create the dish. very nice to use.

- services:

- All the actual logic code is here. For example, if I want to recursively delete all the data in a restaurant, I can write a recursive function here and call it in routes.

- routes:

- The code here is used to handle requests, for example, if I want to create a restaurant, I can write a new route here. as you can see, because we have separated the business logic from the routes, the code here is very clean and very easy to maintain.

- We have written a lot of documentation here , you can see in the api site , including the description of all the request methods , request parameters , return model , error code details and so on.
- We also have a full set of token mechanisms for user authentication. So that the user does not have to log in frequently

My back-end is also responsible for the following things

- Wrote a lot of tests with over 80% coverage.
- We crawled a lot of restaurant data and stored it in the database so that we can display the fake data in the front-end.
